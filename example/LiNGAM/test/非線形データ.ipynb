```python
#Causal Discovery with Cascade Nonlinear Additive Noise
import numpy as np
import csv
import pandas as pd

size = 10000
np.random.seed(123)
N1 = np.random.uniform(size=size, low=-0.5, high=0.5)
np.random.seed(456)
N2 = np.random.uniform(size=size, low=-0.5, high=0.5)
np.random.seed(789)
N3 = np.random.uniform(size=size, low=-0.5, high=0.5)

X1 = N1
X2 = 2*np.tanh(5*X1)+N2
X3 = np.power(X2/2, 3) + N3

xs = np.array([X1, X2, X3]).T
X = pd.DataFrame(np.array([X1, X2, X3]).T,columns=["X1", "X2", "X3"])
X.head()
X.to_csv('nonlinear.csv', index=False)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,8))
plt1 = fig.add_subplot(1, 3, 1)
plt2 = fig.add_subplot(1, 3, 2)
plt3 = fig.add_subplot(1, 3, 3)

plt1.scatter(X1[1:300], X2[1:300])
plt1.set_xlabel('X1')
plt1.set_ylabel('X2')

plt2.scatter(X2[1:300], X3[1:300])
plt2.set_xlabel('X2')
plt2.set_ylabel('X3')

plt3.scatter(X1[1:300], X3[1:300])
plt3.set_xlabel('X1')
plt3.set_ylabel('X3')
```

```python
#https://webdav.tuebingen.mpg.de/causality/NIPS2008-Hoyer.pdf
import numpy as np
import csv
import pandas as pd

size = 10000
np.random.seed(123)
w = np.random.uniform(size=size, low=-3.0, high=3.0)
nx = np.random.uniform(size=size, low=-1.0, high=1.0)
ny = np.random.uniform(size=size, low=-1.0, high=1.0)
nz = np.random.uniform(size=size, low=-1.0, high=1.0)

x=w*w + nx
y=4*np.sqrt(np.abs(w))+ny
z = 2*np.sin(x)+2*np.sin(y) + nz

xs = np.array([w,x,y,z]).T
X = pd.DataFrame(np.array([w,x,y,z]).T,columns=["w", "x", "y", "z"])
X.head()
X.to_csv('nonlinear2.csv', index=False)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,8))
plt1 = fig.add_subplot(1, 4, 1)
plt2 = fig.add_subplot(1, 4, 2)
plt3 = fig.add_subplot(1, 4, 3)
plt4 = fig.add_subplot(1, 4, 4)

plt1.scatter(w[1:300], x[1:300])
plt1.set_xlabel('w')
plt1.set_ylabel('x')

plt2.scatter(x[1:300], y[1:300])
plt2.set_xlabel('x')
plt2.set_ylabel('y')

plt3.scatter(x[1:300], z[1:300])
plt3.set_xlabel('x')
plt3.set_ylabel('z')

plt4.scatter(y[1:300], z[1:300])
plt4.set_xlabel('y')
plt4.set_ylabel('z')
```
